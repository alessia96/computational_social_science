# Plots






```python
import matplotlib.pyplot as plt
from IPython.display import HTML
```

## Wordclouds for each subreddit group

```python
m = plt.imread("wordcloud_mental_nonmental.png")
b = plt.imread("wordcloud_b.png")

fig, ax = plt.subplots(2, figsize=(14, 12), dpi=100)

ax[0].imshow(m)
ax[1].imshow(b)

ax[0].axis("off")
ax[1].axis("off")

plt.show()
```

## Top 25 words in Mental Health and non-Mental health subreddits

```python
plt.figure(figsize=(18, 12), dpi=100)
twm = plt.imread("topwords_ment_vs_nonment.png")
plt.axis("off")
plt.imshow(twm)
```

## Top 25 words in Borderline and Bipolar subreddits
